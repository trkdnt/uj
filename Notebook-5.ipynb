```sql
SELECT egysegar,COUNT(EGYSEGAR) AS 'Milyen eységárak vannak'
FROM Rendeles_tetel
GROUP BY EGYSEGAR
```

```sql
SELECT MAX(Egysegar) AS 'Max', MIN(Egysegar) as 'Min', AVG(Egysegar) AS 'Átlag', COUNT(Egysegar) AS 'Összes rekord (Null érték nélkül)'
FROM Rendeles_tetel
```

```sql
SELECT 163026 + COUNT(*) as Összesen, COUNT(*)-COUNT(EGYSEGAR) as 'Null értekek száma'
FROM rendeles_tetel
WHERE Egysegar IS NULL
```

```sql
SELECT LEN(TERMEKKOD) AS 'karakterek hossza',count(TERMEKKOD) AS 'hány darab ilyen hosszuságú krakterlánc van'
FROM Rendeles_tetel
GROUP BY LEN(TERMEKKOD)
HAVING LEN(TERMEKKOD) IS NOT NULL
```

**1\. Különböző áraknak a gyakorisága**



```sql
SELECT COUNT(list_price) as 'Gyakoriság', list_price
FROM order_items
GROUP BY list_price
order by Gyakoriság 
```

**2\. Max, Min, Átlagos ár és az összes rekord száma**

```sql
SELECT MAX(list_price) AS 'Max', MIN(list_price) as 'Min', AVG(list_price) AS 'Átlag', COUNT(list_price) AS 'Összes rekord'
FROM order_items
```

**3\. A termék ID-nak a hossza és mennyi van belőle**



```sql
SELECT LEN(product_id) AS 'Termék ID hossza',count(product_id) AS 'Darab'
FROM order_items
GROUP BY LEN(product_id)
HAVING LEN(product_id) IS NOT NULL
ORDER BY 'Termék ID hossza'
```

**4\. Mennyi Null érték van**

```sql
SELECT 4722 + COUNT(*) as Összesen, COUNT(*)-COUNT(list_price) as 'Null értekek száma'
FROM order_items
WHERE list_price IS NULL
```
